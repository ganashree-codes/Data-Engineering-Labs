```python
import pandas as pd

#loading the data set into a dataframe

df=pd.read_csv(r"C:\Users\ganas\Desktop\Data Engineering\Input Files\nba.csv")
df

#displaying the data with first 5 rows, number of rows and colums and column name
print("The First five rows : \n",df.head(5))
print("\n")
print("Number of rows and columns : \n",df.shape)
print("\n")
print("All column names: \n",df.columns)
print("\n")

#cleaning the data by converting salary to numeric, removing the null value rows and replacing NaN with unknown values
df['Salary']=pd.to_numeric(df['Salary'],errors='coerce')


df.isnull().sum()

df=df.dropna(subset=['Salary'])

df['College']=df['College'].fillna('Unknown')

#Filtering data based on a requirement
df_team = df[(df["Team"]== "Boston Celtics")]
print("All players from team Boston celtics: \n",df_team)
print("\n")

df_age = df[(df["Age"] > 30)]
print("All players with age greater than 30:\n",df_age)
print("\n")

df_position = df[(df["Position"] == "PG")]
print("All players who plays at PG:\n",df_position)
print("\n")

df_salary=df[(df['Salary'] > 5000000.0)]
print("Players with salary greater than 5000000:\n",df_salary)
print("\n")

#sorting the data
df_highest_salary_player=df.sort_values(by='Salary',ascending=False)
print("All players with highest salary player first:\n",df_highest_salary_player)
print("\n")

df_young_player=df.sort_values(by='Age',ascending=True)
print("All players with youngest first:\n",df_young_player)
print("\n")

#grouping data based on team
df_group=df.groupby('Team')['Salary'].mean()
print("The average salary per team is:\n",df_group)
print("\n")

player_count=df.groupby('Team')['Name'].count()
print("Number of plaers in each team is :\n",player_count)
print("\n")

df_max_salary=df.groupby('Team')['Salary'].max()
print("Maximum salary in each team is :\n",df_max_salary)
print("\n")

#performing a
df_average_age=df['Age'].mean().round(1)
print("The average age of players is ;\n",df_average_age)
print("\n")


df_maximum_salary=df['Salary'].max()
print("The maximum salary among all is :\n",df_maximum_salary)
print("\n")

df_minimum_weight=df['Weight'].min()
print("The minimum weight among all is:\n",df_minimum_weight)
print("\n")


df['Age_in_5_years']=(df['Age'] + 5)
print("The Age of all players after 5 years \n")
print(df)
print("\n")

df['Salary_in_Millions'] = (df['Salary'] / 100000)
print("The Salaries in millions is \n")
print(df)
print("\n")

df_group_highest_salary=df.groupby('Team')['Salary'].sum()
df_group_highest_team = df_group_highest_salary.idxmax()
df_group_highest_salary_maximum=df_group_highest_salary.max()
print("The Team which has the highest salary payout is :\n",df_group_highest_team,df_group_highest_salary_maximum)
print("\n")


df_group_highest_position=df.groupby('Position')['Salary'].mean()
df_group_highest_position_p=df_group_highest_position.idxmax()
df_group_highest_position_salary=df_group_highest_position.max()
print("The position which has highest average salary is :\n",df_group_highest_position_p,df_group_highest_position_salary)
print("\n")


df
```
